```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
import yfinance as yf
```

```python
ticker = ["LSPU.L", "FR0010527275", "CE8.PA", "COMO.PA", "NRJ.PA", "WAT.PA"]
```

```python
stocks = yf.download(tic, period="1y")
```

```python
stocks.head()
```

```python
stocks.to_csv("stocks.csv")
```

```python
stocks = pd.read_csv("stocks.csv", header=[0,1], index_col=[0], parse_dates=[0])
stocks.head()
```

```python
stocks.info()
```

```python
stocks.describe()
```

```python
close = stocks.loc[:, ["Close"]].copy()
```

```python
close.head()
```

```python
close.info()
```

```python
norm = close.div(close.iloc[0]).mul(100)
norm.head()
```

```python
norm.plot(figsize=(15,8))
plt.show()
```

```python
ret = close.pct_change(fill_method=None).dropna()
```

```python
ret.info()
```
