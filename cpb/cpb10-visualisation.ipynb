```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
titanic.plot(subplots=True, figsize=(15, 12), sharex=False)
plt.show()
```

```python
titanic.age.plot(figsize=(12, 8), fontsize=15, c="red", linestyle="-", xlim=(0,900), ylim=(0,80))
plt.title("Titanic - Ages", fontsize=15)
plt.legend(loc=3, fontsize=13)
plt.xlabel("Passanger No", fontsize=12)
plt.ylabel("Age", fontsize=12)
plt.grid()
plt.show()
```

```python
xticks = [x for x in range(0, 901, 50)]
```

```python
yticks = [y for y in range(0, 81, 5)]
```

```python
titanic.age.plot(figsize=(12, 8), fontsize=15, c="red", linestyle="-", xticks=xticks, yticks=yticks, rot=45)
plt.title("Titanic - Ages", fontsize=15)
plt.legend(loc=3, fontsize=13)
plt.xlabel("Passanger No", fontsize=12)
plt.ylabel("Age", fontsize=12)
plt.grid()
plt.show()
```
