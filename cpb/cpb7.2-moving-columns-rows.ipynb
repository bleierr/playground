## Add new column

```python
titanic = pd.read_csv("DataSets/titanic.csv")
```

```python
titanic.head()
```

```python
titanic["Zeros"] = "Zero"
```

```python
titanic.head()
```

```python
titanic.drop(columns="Zeros", inplace=True)
```

```python
titanic.head()
```

```python
#Year of birth
(1912 - titanic["age"])
```

```python
titanic["YoB"] = 1912 - titanic.age
```

```python
titanic.head()
```

```python
titanic["fare_today"] = titanic.fare * 10
```

```python
titanic["total_relatives"] = titanic.sibsp + titanic.parch
```

```python
titanic.head()
```

```python
titanic.drop(columns=["YoB", "fare_today", "total_relatives"], inplace=True)
```

## Insert column at a specific position 

```python
relatives =  titanic.sibsp + titanic.parch
relatives.head()
```

```python
titanic.insert(loc = 6, column = "relatives", value = relatives)
```

```python
titanic.head()
```
