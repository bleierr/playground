## All medals from the 1960s

```python
mask1960s = summer.Year.between(1960, 1969, inclusive="both")
```

```python
og_60s = summer.loc[mask1960s]
og_60s
```

```python
my_favorite_games = [1972,1996]
```

```python
summer.Year.isin(my_favorite_games)
```

```python
og_72_96 = summer.loc[summer.Year.isin(my_favorite_games)]
```

```python
og_72_96.head()
```

```python
#to analyse the opposit use simply the ~ with the same mask
og_not_72_96 = summer.loc[~summer.Year.isin(my_favorite_games)]
```

```python
og_not_72_96.info()
```
